## MODELING 

In this part of our project, we will explore different machine learning methods and come up with the most accurate model for our business problem.

We will write a function that will help us when finding the accuracy, precision recall and f1 scores of our models

```python
# Creating an empty dataframe with column names
scores_df = pd.DataFrame(columns = ['Model', 'Precision', 'Recall', 'Accuracy', 'F1'])

# Writing the function
def enter_scores(Model, y_test, y_hat_test):

    # Calling the global dataframe
    global scores_df
    y_hat_test = Model.predict(X_test)
    # Calculating various test scores
    precision = precision_score(y_test, y_hat_test, average='weighted')
    recall = recall_score(y_test, y_hat_test, average='weighted')
    accuracy = accuracy_score(y_test, y_hat_test)
    f1 = f1_score(y_test, y_hat_test, average='weighted')
    
    # Entering the calculated scores into the dataframe
    scores_df = scores_df.append({'Model':Model, 'Precision':precision, 'Recall':recall, 'Accuracy':accuracy, 'F1':f1}, ignore_index=True)
     #plot confusion matrix for test set:
    plot_confusion_matrix(Model, X_test, y_test, cmap=plt.cm.Blues, normalize='true', xticks_rotation = 'vertical' )
```

Let's separate our categorical and numerical variables 

```python
num = []
cat = []

for i in X.columns:
    if X[i].dtype in ['int64']:
        num.append(i)
    else:
        cat.append(i)
print('Numerical Variables')
print(num)
print('\n')
print('Categorical Variables')
print(cat)
```

### Create Pipeline for Simple Models

We will create pipelines for simple models that will help us form a base that will help up construct even  more complex models

```python
# Create our sub pipelines for numeric and categorial data
s_pipe_num = Pipeline(steps=[('num_imputer', SimpleImputer(strategy = 'median'))])
s_pipe_cat = Pipeline(steps=[('ohe', OneHotEncoder(sparse=False,
                                                    handle_unknown='ignore'))])
s_pipe_smote = SMOTE(sampling_strategy='auto', random_state=42)
```

```python
# Include our sub pipelines in a column transformer to OHE our categorial variables and scale our numerical variables.
preprocessor = ColumnTransformer(transformers=[('s_pipe_num', s_pipe_num, num), 
                                     ('s_pipe_cat', s_pipe_cat, cat)])
preprocessor.fit(X_train)
```

```python
# instantiate each pipeline object for the 3 simpler models
# include column transformer as first step in each pipeline
logreg_pipe = Pipeline(steps=[('preprocessor', preprocessor),
                              ('scaler', MaxAbsScaler()),
                              ('classifier', LogisticRegression(random_state=42, solver='lbfgs', max_iter=150))])

knn_pipe = Pipeline(steps=[('preprocessor', preprocessor), 
                           ('knn', KNeighborsClassifier())])

dtc_pipe = Pipeline(steps=[('preprocessor', preprocessor), 
                           ('dtc', DecisionTreeClassifier(random_state=42))])
```

```python
# Fit each pipeline to our training data
knn_pipe.fit(X_train, y_train)
```

```python
dtc_pipe.fit(X_train, y_train)
```

```python
logreg_pipe.fit(X_train, y_train)
```

```python
#get scores from the logreg simiple model
y_train_pred = logreg_pipe.predict(X_train)
y_test_pred = logreg_pipe.predict(X_test)
enter_scores(logreg_pipe, y_test, y_hat_test)
# Printing out the scores
print('Training Data')
print(classification_report(y_train, y_train_pred ))
print('\n')
print('Testing Data')
print(classification_report(y_test, y_test_pred))
```

```python
y_train_pred = knn_pipe.predict(X_train)
y_test_pred = knn_pipe.predict(X_test)
enter_scores(knn_pipe, y_test, y_test_pred)
# Printing out the scores
print('Training Data')
print(classification_report(y_train, y_train_pred))
print('\n')
print('Testing Data')
print(classification_report(y_test, y_test_pred))
```

```python
y_train_pred = dtc_pipe.predict(X_train)
y_test_pred = dtc_pipe.predict(X_test)
enter_scores(dtc_pipe, y_test, y_test_pred)
# Printing out the scores
print('Training Data')
print(classification_report(y_train, y_train_pred))
print('\n')
print('Testing Data')
print(classification_report(y_test, y_test_pred))
```

Our simple models have the following accuracy scores:
 - Logistic Regression Accuracy: 73%
 - KNearestNeighbors Accuracy: 76%
 - DecisionTree Accuracy: 93%
 
DecisionTree has the highest score of 94%, we suspect that there's an overfitting here. We shall try to solve this as we build on this model.
We shall also develop on the other two models as their intitial scores aren't badly off.

### Random Forest

```python
# Create our random forest pipeline to incorporate column transformer, use default hyperparameters
rfc_pipe = Pipeline(steps=[('preprocessor', preprocessor),
                           ('rfc', RandomForestClassifier(random_state=42))])
# Fit our random forest classifier to the training data
rfc_pipe.fit(X_train, y_train)
```

```python
# Calculate initial scores of random forest with default hyperparameters
y_train_pred = rfc_pipe.predict(X_train)
y_test_pred = rfc_pipe.predict(X_test)
enter_scores(rfc_pipe, y_test, y_test_pred)
# Printing out the scores
print('Training Data')
print(classification_report(y_train, y_train_pred))
print('\n')
print('Testing Data')
print(classification_report(y_test, y_test_pred))
```

Our model has an accuracy of 93% before tuning. It seems there's also some overfitting.
We will try to tune and come up with optimal hyper parameters that will give us the best results under this model.

```python
# create our grid for the grid search parameters
params = {'rfc__n_estimators': [100, 180],
           'rfc__max_depth': [18, 20]}

# set up GridSearchCV object
grid_rfc = GridSearchCV(rfc_pipe, param_grid = params, cv=5)
```

```python
# fit our grid object for random forests to the training data
grid_rfc.fit(X_train, y_train)
```

```python
# find the best hyper parameters from our first grid search
grid_rfc.best_params_
```

```python
params_b = {'rfc__max_depth': [20, 22],
           'rfc__max_features': ['sqrt', 0.2]}

# set up GridSearchCV object
grid_rfc = GridSearchCV(rfc_pipe, param_grid = params_b, cv= 3)
```

```python
# fit our grid object for random forests to the training data
grid_rfc.fit(X_train, y_train)
```

```python
# find the best hyper parameters from our first grid search
grid_rfc.best_params_
```





So for rfc, the best parameters are 'rfc__max_depth': 20, 'rfc__max_features': 0.2 and 'rfc__n_estimators': 100. 
We will model using these hyper parameters and see if there will be an improvement in our scores.

#### Best model for rfc is therefore:

```python
best_params = {'rfc__n_estimators': [200],
           'rfc__max_depth': [20],
           'rfc__max_features': [0.2]}
# set up GridSearchCV object
grid_rfc_best = GridSearchCV(rfc_pipe, param_grid = best_params, cv= 3)
```

```python
grid_rfc_best.fit(X_train, y_train)
```

```python
#Get the scores after including the best parameters
y_train_pred = grid_rfc_best.predict(X_train)
y_test_pred = grid_rfc_best.predict(X_test)
enter_scores(grid_rfc_best, y_test, y_test_pred)
# Printing out the scores
print('Training Data')
print(classification_report(y_train, y_train_pred))
print('\n')
print('Testing Data')
print(classification_report(y_test, y_test_pred))
```

After tuning and using the best hyperparameters, our accuracy score for reduces to 88%, showing that there's overfitting in our model. We will use smote and try fix this issue

#### Best random forest parameters with smote

```python
# Create our random forest pipeline to incorporate column transformer
rfc_best_sm = ImPipeline(steps=[('preprocessor', preprocessor),
                                ('sm', s_pipe_smote),
                                ('rfc', RandomForestClassifier(random_state=42))])
```

```python
# set up GridSearchCV object
grid_rfc_best_sm = GridSearchCV(rfc_best_sm, param_grid = best_params, cv = 3)
```

```python
grid_rfc_best_sm.fit(X_train, y_train)
```

```python
#Scores after applying smote

y_train_pred = grid_rfc_best_sm.predict(X_train)
y_test_pred = grid_rfc_best_sm.predict(X_test)
enter_scores(grid_rfc_best_sm, y_test, y_test_pred)
# Printing out the scores
print('Training Data')
print(classification_report(y_train, y_train_pred))
print('\n')
print('Testing Data')
print(classification_report(y_test, y_test_pred))
```

After using smote, the accuracy score dropped to 86%. This has helped reduce the overfitting problem.

### Combining XGBoost with RF using Stacking

```python
# Instantiate initial stacking classifier object with our two models
x_r_stack = StackingClassifier(estimators=[('RandForest', grid_rfc_best),
                                           ('XGBoost', grid_b_p_XGB)])
```

```python
# Fit the stacking classifier to our training data
x_r_stack.fit(X_train, y_train)
```

```python
#stack  scores 
y_train_pred = x_r_stack.predict(X_train)
y_test_pred = x_r_stack.predict(X_test)
enter_scores(x_r_stack, y_test, y_test_pred)
# Printing out the scores
print('Training Data')
print(classification_report(y_train, y_train_pred))
print('\n')
print('Testing Data')
print(classification_report(y_test, y_test_pred))
```

This model combines the two models that have given us the best scores so far. 
It has an accuracy of 87%
