##                BUSINESS PROBLEM

Despite having access to enormous volumes of fresh water, many Tanzanians still experience water shortages because of inadequate infrastructure for accessing and storing it in both rural and urban regions. Only a small number of homes have access to clean, piped water for drinking. Only a small percentage of rural households have access to water for agriculture irrigation.

A combination of measures on the systems and hard infrastructure are needed to improve access to water. In both urban and rural locations, pipes, irrigation systems, and pumps must be built. Thankfully, the government and other well-wishers acted and substantially expanded public resources to the water industry. To supply water to as many homes as feasible, wells and pumps have been constructed.



The Tanzanian Ministry of Water has recruited this data science consulting organization to develop a model to help predict whether water pumps are functioning, functional but in need of repairs, or non-functional.

This model will help reduce wastage of the Ministry's resources by only dispatching repair teams to malfunctioning or potentially in need of repair pumps. In addition to maximizing accuracy, we also want to increase recall to make sure Tanzanians have access to potable water and that few malfunctioning or damaged pumps go unnoticed.

### EXPLORATORY DATA ANALYSIS

###### The distribution of wells in various regions 

```python
#plot a countplot to show how pumps are distributed in various regions
plt.figure(figsize=[24,14])
ax = sns.countplot(data=df, x="region", hue="status_group", orient = 'h')
plt.xticks(size=16, rotation = 15, ha="right")
plt.yticks(size=16)
plt.title("Well Distribution", size= 20)
```

 - Iringa has the highest number of functional pumps followed by Shinyanga  and Kilimanjaro.
 - Mbeya, Morogoro and Shinyanga lead in the regions that have non functional pumps.
 - Shinyanga, Kigoma and Mbeya are the top three regions that have pumps that are functional but need repairs.


##### Investingating the relationship between various features and the target

```python
df.columns
```

```python
#plot a countplot to show how pumps are distributed according to the modes of extraction
plt.figure(figsize=[24,14])
ax = sns.countplot(data=df, x = 'extraction_type_class', hue = "status_group", palette='muted')
plt.xticks(size = 18, rotation = 15, ha="right")
plt.yticks(size = 18)
plt.title("Well Distribution", size= 20)
```

Wells where gravity is used to extract water seem to have the highest number of pumps that are functional. It's also noticable that this has the highest number of non functional pumps.
Motorpump and 'other' methods of extraction are interesting since they have a higher number of those pumps that are non functional than functional. This means that these method of extraction need to be investaged so that it can be known why there is a high rate of failure.

```python
#bar plots showing the distribution of quantity groups
df["quantity_group"].value_counts().plot(figsize = [10,8], kind="bar", color = "green")
plt.xlabel("quantity groups")
plt.ylabel("Count")
plt.title("Distribution of Quantity groups")
plt.legend()
```

Most water quantity distributions is enough. Though there's a big percentage of insufficient. Dry and seasonal also have signifacant portions of the data.

```python
#plot a countplot showing different managements and there pump status distributions
plt.figure(figsize=[24,14])
ax = sns.countplot(data=df, x = 'management', hue = "status_group", palette='dark')
plt.xticks(size = 18, rotation = 15, ha="right")
plt.yticks(size = 18)
plt.title("Well Distribution", size= 20)
```

VWC manages the highest number of wells. This is also trickles down to the status of the pumps as it leads in each category ahead of other management.

##### Investigating how the target categories are distributed

```python
y = df.status_group
y.value_counts(normalize=True)
```

```python
#draw graph to show us the distribution of the target categories
fig, ax = plt.subplots(figsize = (10,8))
plt.bar(x = y.unique(), height = round(y.value_counts(normalize=True)*100,2))
plt.xticks(ticks=[0,1,2], labels =['functional','non functional','functional needs repair'], fontsize = 10)
plt.yticks(fontsize = 10)
ax.set_xlabel('Functionality', fontsize = 15)
ax.set_ylabel('Percentage', fontsize = 15)
plt.suptitle('Well Status', fontsize = 15)
plt.show()
```

Generally, most water pumps are in a functional state. 38% of the pumps in Tanzania however are non functional. Only &% of the pumps need repairs so that they can start functioning well
